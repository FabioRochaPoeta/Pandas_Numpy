# Biblioteca Pandas com foco nas aplicações em Ciência de Dados

O Pandas é uma biblioteca open source para análise de dados em Python criada em 2008, que oferece ferramentas para manipular, analisar e visualizar dados de forma rápida e fácil. Com o Pandas, você pode carregar, limpar, transformar e mesclar seus dados com apenas algumas linhas de código de forma  rápida, poderosa, flexível e fácil de usar.

## Na Ciência de Dados, o Pandas é uma ferramenta essencial. Alguns exemplos de usos comuns do Pandas incluem:

- **Limpeza e pré-processamento de dados:** o Pandas pode ser usado para limpar e transformar dados, incluindo a remoção de valores nulos e duplicados.

- **Análise exploratória de dados:** o Pandas oferece recursos para explorar dados e encontrar insights, como cálculos estatísticos e visualizações de dados.

- **Manipulação de dados em tempo real:** o Pandas pode ser usado para processar e analisar dados em tempo real, incluindo streaming de dados de sensores e dispositivos.

- **Preparação de dados para treinamento de modelos:** o Pandas é frequentemente usado para preparar dados para o treinamento de modelos de Machine Learning, incluindo o ajuste de escalas e transformações de dados.

- **Análise de dados de redes sociais:** o Pandas pode ser usado para analisar dados de redes sociais, incluindo a extração e análise de tweets, posts de blog e outras fontes de dados não estruturados.

- **Ferramentas para ler e gravar dados entre estruturas de dados na memória e diferentes formatos:** CSV e arquivos de texto, Microsoft Excel, bancos de dados SQL e o formato rápido HDF5;

- **Análise de dados de pesquisa:** o Pandas pode ser usado para analisar dados de pesquisa, incluindo dados de pesquisas de opinião pública e de comportamento do consumidor.

### As duas principais estruturas de dados do Pandas

**Series**

Series é uma matriz rotulada unidimensional capaz de conter qualquer tipo de dados (inteiros, strings, números de ponto flutuante, objetos Python, etc.) 

Os rótulos dos eixos são chamados coletivamente de índice. O método básico para criar uma série é:

![alt](https://pandas.pydata.org/docs/_images/01_table_series.svg)

```python
 s1 = pd.Series(['a','b','c','d'], index= [1,2,3,4], name='Serie1')
s1
```

```python
 s2 = pd.Series(['e','f','g','h'], index= [1,2,3,4], name='Serie2')
s2
```

**Dataframes**

DataFrame é uma estrutura de dados rotulada bidimensional com colunas de tipos potencialmente diferentes. Você pode pensar nisso como uma planilha ou tabela SQL, ou um dicionário de objetos Series. 

Geralmente é o objeto pandas mais comumente usado.
Como o Series, o DataFrame aceita muitos tipos diferentes de entrada:

- Dict of 1D ndarrays, listas, dicts ou Series 
- 2-D numpy.ndarray
- Ndarray estruturado ou de registro
- UMA Series
- Outro DataFrame

![alt](https://pandas.pydata.org/docs/_images/01_table_dataframe.svg)

```python
pd.DataFrame([s1,s2])
```

```python
pd.DataFrame([s1,s2]).T
#Transpor
```

```python
products[(products["product_name_lenght"] == 20) & (products["product_category_name"] == "perfumaria")]
#Filtrando por dois valores específicos de duas colunas
```

### Operações:

- unique
- mean
- apply
- value_counts
- sort_values

```python
products["product_category_name"].unique()
#Valores únicos de uma coluna
```

```python
products["product_name_lenght"].mean()
#Média de uma tabela específica
```

```python
products["product_category_name"].value_counts()
```

```python
(products["product_category_name"] == "perfumaria").value_counts()
#Contando os registros de um valor específico de uma coluna específica
```

```python
(products["product_category_name"].value_counts()).sort_values(ascending=True)
```

```python
products.sort_values(by=["product_name_lenght"],ascending=True)
```
